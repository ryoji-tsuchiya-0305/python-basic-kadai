```python
array = ["水","金","地","火","木","土","天","海","冥"]

for index,value in enumerate(array):

  print(f" {index+1}  {value}")

index = 0
value = array[0]

while index < 9:

  value = array[index]
  print(f"{index+10}  {value}")
  index = index + 1
```
